# Exercise 2.1

Analyze the writing patterns of each student

```python
# Define una función para analizar el patrón de escritura de un texto
def analizar_patron_escritura(texto):
    if pd.notna(texto) and texto.strip() != '#N/D':  # Validar que la celda contenga texto y no '#N/D'
        # Tokenizar el texto en palabras
        palabras = word_tokenize(texto)
        lemmatizer = WordNetLemmatizer()

        # Eliminar stopwords (palabras comunes que no aportan información)
        stopwords_espanol = set(stopwords.words("spanish"))
        stopwords_english = set(stopwords.words("english"))
        palabras_filtradas = [lemmatizer.lemmatize(palabra.lower()) for palabra in palabras if palabra.isalnum() and palabra.lower() not in stopwords_espanol and palabra.lower() not in stopwords_english]

        # Calcular la frecuencia de palabras
        frecuencia_palabras = FreqDist(palabras_filtradas)

        # Mostrar las palabras más comunes
        print("Palabras más comunes en la respuesta:")
        print(frecuencia_palabras.most_common(10))

# Iterar a través de los estudiantes (excluyendo la primera columna) y agrupar todas las respuestas
for indice, fila in df.iloc[:, 1:].iterrows():  # Excluir la primera columna
    respuesta_agrupada = " ".join(fila.dropna())  # Unir todas las respuestas de las columnas sin valores nulos
    print(f"Análisis para el estudiante {indice + 1}:")
    analizar_patron_escritura(respuesta_agrupada)
    print(f"Longitud de la respuesta agrupada: {len(respuesta_agrupada)} caracteres\n")
    #print(f"Cantidad de palabras de respuesta agrupada: {count(respuesta_agrupada)} caracteres\n")
```

# Exercise 2.2

Evaluate the similarities of the homeworks of the students.

At a homework level, then as a student level.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
def preprocesar_y_agrupar_respuestas(fila):
    respuestas = [str(fila[columna]) for columna in df.columns[1:] if pd.notna(fila[columna]) and fila[columna].strip() != '#N/D']
    return ' '.join(respuestas)

# Preprocesar y agrupar respuestas de los estudiantes
respuestas_agrupadas = df.apply(preprocesar_y_agrupar_respuestas, axis=1)

# Calcular la matriz TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(respuestas_agrupadas)

# Calcular la similitud coseno entre todos los estudiantes
similitud_matrix = cosine_similarity(tfidf_matrix)

# Crear un DataFrame para mostrar la similitud en forma de tabla
similitud_df = pd.DataFrame(similitud_matrix, columns=df.index, index=df.index)

# Mostrar la tabla de similitud
print(similitud_df)
```
